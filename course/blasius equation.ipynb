```python
import numpy as np
import matplotlib.pyplot as plt

# init conditions
f0 = [0]
df = [0]
d2f = [0.3]

dn = 0.5
max_n = 6
n = np.linspace(0,max_n,int(max_n/dn)+1)

for i in range(len(n)-1):
    n_df = df[-1] + dn*d2f[-1]
    n_f = f0[-1] + dn*df[-1]
    n_d2f = d2f[-1] - 0.5*d2f[-1]*f0[-1]
    f0.append(n_f)
    df.append(n_df)
    d2f.append(n_d2f)

# print result
print (" "*2+ 'n'+ " "*4 +'f' + " "*4 + 'f \' ' + " "*3 + 'f\" ')
print('-'*23)
for i in range(len(n)):
    print(f'{n[i]} {f0[i]:.3f} {df[i]:.3f} {d2f[i]:.3f}')


# visualization boundary layer
plt.figure()
plt.plot(n,df)
plt.show()
```
